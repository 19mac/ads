```python
#Mean Method
fds1 = pd.read_csv("/content/fashion_products.csv")
fds1.head()
```

```python
fds1.isnull().sum()
```

```python
missing_col = ["User ID"]
#Using mean to impute the missing values
for i in missing_col:
    fds1.loc[fds1.loc[:,i].isnull(),i]=fds1.loc[:,i].mean()
fds1.isnull().sum()
```

```python
fds1.isnull().sum()
```

```python
#Median Method
fds2 = pd.read_csv("/content/fashion_products.csv")
fds2.head()
```

```python
fds2.isnull().sum()
```

```python
missing_col_1 = ["Product ID"]
#Using mean to impute the missing values
for i in missing_col_1:
    fds2.loc[fds2.loc[:,i].isnull(),i]=fds2.loc[:,i].median()
fds2.isnull().sum()
```
